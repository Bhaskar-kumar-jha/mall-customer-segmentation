```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
```

```python
import pandas as pd

url = "https://raw.githubusercontent.com/gakudo-ai/open-datasets/refs/heads/main/Mall_Customers.csv"
df = pd.read_csv(url)
df.head()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
plt.scatter(df['Annual Income (k$)'], df['Spending Score (1-100)'], c='blue', edgecolors='black')
plt.title('Customer Segments')
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.grid(True)
plt.show()
```

```python
from sklearn.cluster import KMeans

# Prepare the features
X = df[['Annual Income (k$)', 'Spending Score (1-100)']]

# Create the KMeans model with 5 clusters (you can change this number later if needed)
kmeans = KMeans(n_clusters=5, random_state=42)
kmeans.fit(X)

# Add the predicted cluster labels to the dataframe
df['Cluster'] = kmeans.labels_
```

```python
import matplotlib.pyplot as plt

# Set figure size
plt.figure(figsize=(8, 6))

# Scatter plot with color-coded clusters
plt.scatter(X['Annual Income (k$)'], X['Spending Score (1-100)'],
            c=kmeans.labels_, cmap='rainbow', s=100)

# Add cluster centers
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], color='black', marker='X', s=200, label='Centroids')

# Add labels and title
plt.xlabel('Annual Income (k$)')
plt.ylabel('Spending Score (1-100)')
plt.title('Customer Segments - KMeans Clustering')
plt.legend()
plt.grid(True)
plt.show()
```
